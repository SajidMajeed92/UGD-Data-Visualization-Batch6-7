```python
# write a function which return a team name the players scored most runs against that team
def batsman_run(bats_man):
    team=delivery[delivery['batsman']==bats_man]
    return team.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).index[0]
```

```python
batsman_run('DA Warner')
```

```python
delivery.groupby('batsman')['batsman_runs'].count()
```

```python
death_overs=delivery[delivery['over']>15]
```

```python
death_overs.groupby('batsman')['batsman_runs'].count()>200
```

```python
all_batsman=death_overs.groupby('batsman')['batsman_runs'].count()
x= all_batsman>200
batsman_list = all_batsman[x].index.tolist()
```

```python
batsman_list
```

```python
final=delivery[delivery['batsman'].isin(batsman_list)]
```

```python
runs=final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
strike_rate= (runs/balls)*100
```

```python
strike_rate.sort_values(ascending=False).head()
```
